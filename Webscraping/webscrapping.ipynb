# WebScrapping

## El objetivo de este webscrapping consiste en conseguir información de todas las cartas que necesitamos, de manera automática y lo más sencillo posible
## Los datos que necesitamos se clasifican en dos grupos:

* ### Imagenes de la carta
* ### Datos básicos de la carta (Nombre, Descripción, Color, Número)

```python
# Este primer script aporta el primer grupo de datos que necesitamos, que son imágenes

import json
from urllib.request import Request, urlopen

def get_image(cardnumber):
    req = Request(
        url='https://images.digimoncard.io/images/cards/{number}.jpg'.format(number = cardnumber), 
        headers={'User-Agent': 'Mozilla/5.0'}
    )
    image = urlopen(req).read()
    with open('./img/{number}.jpg'.format(number = cardnumber), 'wb') as handler:
        handler.write(image)

with open("cards.json") as content:
    cards = json.load(content)
    for card in cards:
        get_image(card.get('cardnumber'))
```
